```python
sentence = 'The cat chased all dogs down the street.'
```

```python
with corenlp.CoreNLPClient(start_server=False) as client:
    ann = client.annotate(sentence)
```

```python
ann
```

```python
ann.sentence[0]
```

```python
for edge in ann.sentence[0].collapsedDependencies.edge:
    # see http://universaldependencies.org/docsv1/u/dep/nsubj.html
    if "nsubj" in edge.dep:
        print("==============\n nominal subject:\n", edge)
        source = edge.source
        source_node = ann.sentence[0].collapsedDependencies.node[source]
        target = edge.target
        target_node = ann.sentence[0].collapsedDependencies.node[target]
        print("Source:", ann.sentence[0].token[source_node.index-1])
        print("Target:", ann.sentence[0].token[target_node.index-1])
```

```python
# The is_passive function accepts 
# Checks in ''nsubjpass'' appears in the dependency parsing of the sentence
# and returns True / False respectively

def is_passive(sentence):
    dep_parse = nlp.dependency_parse(sentence)
    # Check if the nsubjpass appears in the dependencies
    check = [t for t in dep_parse if 'nsubjpass' in t]
    # Returns True if the length of the list check is greater than zero
    # (i.e., the nsubjpass appears in there)
    return len(check)>0
```

```python
passive = is_passive(sentence)
passive
```

```python
tokens = nlp.word_tokenize(sentence)
tokens
```
